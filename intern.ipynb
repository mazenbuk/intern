#Univariate Data Analysis

```python
age_stats = data['Age'].describe()
age_stats
```

##Analisis untuk variabel 'Age' memberikan informasi berikut:

- Jumlah data (count): 1000
- Rata-rata (mean): 37.17 tahun
- Standar deviasi (std): 12.01 tahun, menunjukkan seberapa lebar variasi umur dalam dataset.
- Nilai minimum (min): 14 tahun
- Kuartil pertama (25%): 27.75 tahun, artinya 25% dari data berada di bawah umur ini.
- Median (50%): 36 tahun, artinya setengah dari data memiliki umur kurang dari 36 tahun dan setengah lainnya lebih.
- Kuartil ketiga (75%): 45 tahun, artinya 75% dari data berada di bawah umur ini.
- Nilai maksimum (max): 73 tahun

```python
plt.figure(figsize=(12, 8))
sns.histplot(data['Age'], kde=True)
plt.title("Age")
plt.show()
```

```python
gender_frequency = data['Gender'].value_counts()
gender_frequency
```

##Analisis untuk variabel 'Gender' menunjukkan frekuensi untuk setiap kategori sebagai berikut:

- Kategori 1: 598 entri, yang mewakili gender laki-laki.
- Kategori 2: 402 entri, yang mewakili gender perempuan.

```python
labels = gender_frequency.index.map({1: 'Laki-Laki', 2: 'Perempuan'})
sizes = gender_frequency.values
colors = ['#ff9999','#66b3ff']
explode = (0.1, 0)

plt.figure(figsize=(8, 6))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('Gender')
plt.show()
```

#Bivariate Data Analysis

```python
plt.figure(figsize=(10, 6))
plt.title('Distribution of Smoking by Disease Level')
sns.boxplot(x='Level', y='Smoking', data=data, order=['Low', 'Medium', 'High'])
plt.xlabel('Disease Level')
plt.ylabel('Smoking')
plt.show()
```

##Boxplot di atas menunjukkan distribusi 'Smoking' untuk setiap 'Level' penyakit (Low, Medium, High).

- Terdapat variasi dalam tingkat merokok di antara ketiga kelompok level penyakit.
- Kelompok dengan level penyakit "High" tampak memiliki median tingkat merokok yang cenderung lebih tinggi dibandingkan dengan kelompok "Low" dan "Medium"
- Jangkauan interkuartil dan outliers menunjukkan variasi dalam tingkat merokok di setiap kategori level penyakit, menandakan bahwa individu dalam masing-masing kategori memiliki kebiasaan merokok yang beragam.
